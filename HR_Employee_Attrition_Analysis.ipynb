```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\Varun\OneDrive\Desktop\WA_Fn-UseC_-HR-Employee-Attrition.csv")
```

```python
print("‚úÖ First 5 Rows:")
display(df.head())  
```

```python
print("\nüîç Dataset Info:")
df.info()  #dataset info
```

```python
print("\n Missing Values:")
print(df.isnull().sum())  #missing values
```

```python
# Count of Attrition values
attrition_counts = df['Attrition'].value_counts()

# Pie chart
plt.figure(figsize=(6,6))
plt.pie(attrition_counts, labels=attrition_counts.index, autopct='%1.1f%%', startangle=90, colors=['#66c2a5','#fc8d62'])
plt.title('Overall Employee Attrition Rate')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(data=df, x='Gender', hue='Attrition')
plt.title("Attrition by Gender")
plt.ylabel("Number of Employees")
plt.show()
#attrition of gender
```

```python
plt.figure(figsize=(9,5))
sns.countplot(data=df, x='Department', hue='Attrition')
plt.title("Attrition by Department")
plt.xticks(rotation=20)
plt.ylabel("Number of Employees")
plt.show()
#attrition of department
```

```python
plt.figure(figsize=(10, 5))
sns.countplot(data=df, x='JobRole', hue='Attrition')
plt.title('Attrition by Job Role')
plt.xlabel('Job Role')
plt.ylabel('Number of Employees')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
#attrition by job role
```

```python
plt.figure(figsize=(6, 5))
sns.countplot(data=df, x='OverTime', hue='Attrition')
plt.title('Attrition by Overtime Work')
plt.xlabel('Overtime')
plt.ylabel('Number of Employees')
plt.tight_layout()
plt.show()
#attrition by overtime
```

```python
plt.figure(figsize=(6, 5))
sns.boxplot(data=df, x='Attrition', y='MonthlyIncome')
plt.title('Monthly Income Distribution by Attrition')
plt.xlabel('Attrition')
plt.ylabel('Monthly Income')
plt.tight_layout()
plt.show()
#monthly income vs attrition
```

```python
plt.figure(figsize=(8, 5))
sns.kdeplot(data=df[df['Attrition'] == 'Yes']['Age'], label='Left Company', fill=True)
sns.kdeplot(data=df[df['Attrition'] == 'No']['Age'], label='Stayed', fill=True)
plt.title('Age Distribution by Attrition')
plt.xlabel('Age')
plt.legend()
plt.tight_layout()
plt.show()
#Age Distribution by Attrition
```
